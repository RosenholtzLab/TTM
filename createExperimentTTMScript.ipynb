## Some files failed for Foveated Case. Figure out which oens and resbumit

```python
#create map of fixation location to eccentricity
import numpy as np
bbx_key_80 = np.array((bbx_list['image_name'],
                       bbx_list['fx_x_16_5'].round().astype(int),
                       bbx_list['fx_y_16_5'].round().astype(int))).T
bbx_key_160 = np.array((bbx_list['image_name'],
                        bbx_list['fx_x_16_10'].round().astype(int),
                        bbx_list['fx_y_16_10'].round().astype(int))).T
bbx_key_240 = np.array((bbx_list['image_name'],
                        bbx_list['fx_x_16_15'].round().astype(int),
                        bbx_list['fx_y_16_15'].round().astype(int))).T

#move output files to new folder
new_folder_name = '/home/gridsan/groups/RosenholtzLab/foveated_exp_stims'
experiment_folder_name = '/home/gridsan/groups/RosenholtzLab/cocop_stims'
import glob
output_folder = "/home/gridsan/vdutell/TTM/foveated_output/*/mongrelized_*.png"
mongrel_paths = glob.glob(output_folder)
fnames = []
fx_xs = []
fx_ys = []
inpaint_flag = []
for path in mongrel_paths:
    ecc=None
    command=None
    parts = path.split('/')
    folder = parts[6]
    fname = parts[7]
    folderparts = folder.split('_')
    imnum = folderparts[0]
    fnames.append(imnum)
    
    if(folderparts[1]=='inpaint'):
        inpaint_flag.append('True')
        fx = folderparts[2].replace('X','')
        fy = folderparts[3].replace('Y','')
        #command = f'cp {path} {new_folder_name}/absent/{imnum}_{fx}_{fy}.png'
        #also copy to final experiment folder
        #check eccentricity
        mytup = np.array((f'{imnum}.jpg',int(fx),int(fy)), dtype=object)
        #print(mytup)
        #print(bbx_key_80[0])
        if (mytup == bbx_key_80).all(axis=1).any():
            ecc = 80
        elif (mytup == bbx_key_160).all(axis=1).any():
            ecc = 160
        elif (mytup == bbx_key_240).all(axis=1).any():
            ecc = 240
        if ecc==None:
            print(f'Failed to find target absent image {imnum}, fx{fx}, fy{fy}')
        else:
            command = f'cp {path} {experiment_folder_name}/foveated/absent/{ecc}/{imnum}_{ecc}_foveated_absent.png'
            #print(f'Found ecc {ecc}')
    else:
        inpaint_flag.append('False')
        fx = folderparts[1].replace('X','')
        fy = folderparts[2].replace('Y','')
        #command = f'cp {path} {new_folder_name}/present/{imnum}_{fx}_{fy}.png'
        #check eccentricity
        mytup = np.array((f'{imnum}.jpg',int(fx),int(fy)), dtype=object)
        #print(mytup)
        if (mytup == bbx_key_80).all(axis=1).any():
            ecc = 80
        elif (mytup == bbx_key_160).all(axis=1).any():
            ecc = 160
        elif (mytup == bbx_key_240).all(axis=1).any():
            ecc = 240
        #also copy to final experiment folder
        if ecc==None:
            print(f'Failed to find target present image {imnum}, fx{fx}, fy{fy}')
        else:
            command = f'cp {path} {experiment_folder_name}/foveated/present/{ecc}/{imnum}_{ecc}_foveated_present.png'
            #print(f'Found ecc {ecc}')
        
    #print(command)
    os.system(command)
#print(fnames[:5],fx_xs[:5],fx_ys[:5],inpaint_flag[:5])
```

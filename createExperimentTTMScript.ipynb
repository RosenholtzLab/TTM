## Create Foveated TTM Target Present Files

```python
folder = '/home/gridsan/groups/RosenholtzLab/cocop_stims/'
ppd=16
#script for uniform TTM

#target present
count = 0;
list_count = 1;
# with open('ttm_present_list.txt','w') as file:

output_dir = 'ttm_present_lists'
os.makedirs(output_dir,exist_ok=True)
list_name = f'ttm_present_{list_count}.txt'
text_file = open(os.path.join(output_dir,list_name), "w")
for imname in bbx_list['image_name']:
    for deg_ecc in [5,10,15]:
        fx_x = bbx_list.loc[bbx_list['image_name'] == imname, f'fx_x_{ppd}_{deg_ecc}'].item()
        fx_y = bbx_list.loc[bbx_list['image_name'] == imname, f'fx_y_{ppd}_{deg_ecc}'].item()
        fx_x = np.round(fx_x).astype(int)
        fx_y = np.round(fx_y).astype(int)
        line = f'{folder}{imname}\t{fx_x}\t{fx_y}\t{ppd}\t{count}\n'
        text_file.write(line)
        count += 1
        if count%23==0:
            text_file.close()
            list_count += 1
            list_name = f'ttm_present_{list_count}.txt'
            text_file = open(os.path.join(output_dir,list_name), "w")
text_file.close()
            
        
        

```

## Create Foveated TTM Target Absent Files

```python
count = 0;
list_count = 1;
# with open('ttm_present_list.txt','w') as file:

output_dir = 'ttm_absent_lists'
os.makedirs(output_dir,exist_ok=True)
list_name = f'ttm_absent_{list_count}.txt'
text_file = open(os.path.join(output_dir,list_name), "w")
for imname in bbx_list['image_name']:
    im_num = imname.replace('.jpg','')
    im_inpaint = f'{im_num}_inpaint.png'
    for deg_ecc in [5,10,15]:
        fx_x = bbx_list.loc[bbx_list['image_name'] == imname, f'fx_x_{ppd}_{deg_ecc}'].item()
        fx_y = bbx_list.loc[bbx_list['image_name'] == imname, f'fx_y_{ppd}_{deg_ecc}'].item()
        fx_x = np.round(fx_x).astype(int)
        fx_y = np.round(fx_y).astype(int)
        line = f'{folder}{im_inpaint}\t{fx_x}\t{fx_y}\t{ppd}\t{count}\n'
        text_file.write(line)
        count += 1
        if count%23==0:
            text_file.close()
            list_count += 1
            list_name = f'ttm_absent_{list_count}.txt'
            text_file = open(os.path.join(output_dir,list_name), "w")
text_file.close()
```

## Create Uniform TTM Target Absent Files

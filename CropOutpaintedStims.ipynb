## Readin Images, Get Sizes, and Crop!

```python
import glob
import cv2
imfolder = '/home/gridsan/vdutell/RosenholtzLab_shared/cocop_stims'
imlist = glob.glob(f'{imfolder}/*.jpg', recursive=False)
len(imlist)
```

```python
masklist = []
for impath in imlist:
    #read in original to get size and imnumber
    im = cv2.imread(impath)
    imnum = impath.split('/')[-1].replace('.jpg','')
    imsize = np.shape(im)
    print(imsize)
    print(impath)
    #read in inpainted ( +outpainted) image
    old_inpaint_path = impath.replace('cocop_stims','cocop_stims/outpainted').replace('.jpg','_inpaint.png')
    print(old_inpaint_path)
    iminpaint = cv2.imread(old_inpaint_path)
    old_inpaint_size = np.shape(iminpaint)
    #crop to original size
    iminpaint = iminpaint[:imsize[0],:imsize[1],:imsize[2]]
    print(iminpaint.shape)
    #save inpainted image
    savepath = f'/home/gridsan/groups/RosenholtzLab/cocop_stims/{imnum}_inpaint.png'
    cv2.imwrite(savepath,iminpaint,[int(cv2.IMWRITE_PNG_COMPRESSION),0])
    #print(noisemask.shape)
    #plt.imshow(noisemask)
    #plt.title(imnum)
    #plt.show()
    print('***')
```
